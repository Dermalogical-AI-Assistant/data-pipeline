```python
df = initial_product_all_backup[pd.to_datetime(initial_product_all_backup['collected_day']).dt.month == 4]
df.loc[:, 'price'] = df.loc[:, 'price'].str.replace("£", "")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace(r"[\[\]']", "", regex=True)
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("acne-blemishes", "ACNE_BLEMISHES")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("anti-ageing", "ANTI_AGING")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("sensitive", "SENSITIVE_SKIN")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace('normal-combination', "COMBINATION_SKIN")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("redness-rosacea", "REDNESS")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("oily-skin", "OILY_SKIN")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("dull-skin", "DULL_SKIN")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("dry-skin", "DRY_SKIN")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("barrier-repair", "DAMAGED_SKIN_BARRIER")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("dark-circles", "DARK_CIRCLES")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace('uneven-skintone', "UNEVEN_SKINTONE") 
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("scars", "PIGMENTATION")
df.loc[:, 'skincare_concern'] = df.loc[:, 'skincare_concern'].str.replace("blocked-pores-dead-skin", "BLACKHEADS_PORES")
df = df.drop(columns=['_id', 'url', 'collected_day'])
df = df.rename(columns={
    'img': 'thumbnail', 
    'full_ingredients_list': 'fullIngredientsList', 
    'how_to_use': 'howToUse', 
    'ingredient_benefits': 'ingredientBenefits',
    'skincare_concern': 'skincareConcerns'
})
df
```
